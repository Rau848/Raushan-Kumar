Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
import plotly.graph_objects as go
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=titanic.age,y=titanic.fare,mode='markers'))
fig.show()
```

Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
import plotly.express as px

tips_data = px.data.tips()
fig = px.box(tips_data, x='day', y='total_bill', color='sex')
fig.show()
```

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
tips_data = px.data.tips()
fig = px.histogram(tips_data, x='sex', y='total_bill', color='day', pattern_shape='smoker')
fig.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.

```python
iris = px.data.iris()
fig = px.scatter_matrix(iris, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"], color="species")
fig.show()
```

5. What is Distplot? Using Plotly express, plot a distplot.

The distplot function in Plotly Express is used to create a histogram with an overlaid kernel density estimation (KDE) plot. It provides a visual representation of the distribution of a single variable.

```python
import plotly.express as px
data = px.data.tips()
fig = px.histogram(data, x="total_bill", nbins=30, marginal="rug", opacity=0.7)
fig.update_layout(title="Distribution of Total Bill",xaxis_title="Total Bill",yaxis_title="Density")
fig.show()
```
